# **TASK 2- DATA CLEANING AND PRE-PROCESSING**

The objective of this task is to clean,standardise and transform the HR dataset to ensure accuracy, consistency and suitability for EDA.
This includes handling data types, standardising column names, managing missing values, removing duplicates, and creating new meaningful features

**1. Import required libraries**

- In this step, we import necessery Python libraries required for data analysis and manipulation

```python
import pandas as pd
import numpy as np
```

**2. Load the dataset**

- The HR dataset is loaded using pandas to begin the preprocessing phase.
- The dataset was stored locally in the same directory as the Jupyter Notebook and loaded using Pandas.

```python
df = pd.read_excel("HR_Main_project_dataset.xlsx")
```

**3. Initial data exploration**

- This step helps in understanding the structure, size and data types of the dataset

```python
df.head()
df.shape
df.info()
df.describe()
```

This step helps identify numerical, categorical and date variables as well as potential data quality issues.

**Sort rows by employee ID**

```python
df=df.sort_values(by="EmpID")
```

- Employee records are sorted by employee ID to improve data readability.

**4. Checked the shape**

```python
print("Number of rows:",df.shape[0])
```

```python
print("Number of columns:",df.shape[1])
```

**5. Checking for unique values**

```python
df["EmpID"].nunique()
```

- A uniqueness chaeck was performed on the employee ID to confirm that each row represents a unique employee record. 

**6. Checking for missing values**

- Missing values are examined across all columns.

```python
df.isnull().sum()
```

- The DateofTermination column contains null values for employees who are currently active in the organization.
These null values are valid and expected, as termination dates apply only to employees who have exited the company.
Therefore, no imputation or deletion was performed for this column.

**7. Handling missing values**

**7.1 Date Columns**
- Date related columns are converted to datetime format to support time based analysis.
- Missing termination dates imply active employees and missing values are handled safely

```python
df["DateofTermination"]=pd.to_datetime(df["DateofTermination"],errors="coerce")
df["LastPerformanceReview_Date"]=pd.to_datetime(df["LastPerformanceReview_Date"],errors="coerce")
```

- invalid or missing dates are safely converted to NaT

**7.2 Categorical Columns**
- Missing values are replaced with "Unknown"

```python
cat_columns=df.select_dtypes(include="object").columns
df[cat_columns]=df[cat_columns].fillna("Unknown")
```

**7.3 Numerical columns**

- Missing values are filled with Median

```python
num_columns=df.select_dtypes(include=["int64","float64"]).columns
df[num_columns]=df[num_columns].fillna(df[num_columns].median())
```

**8. Removal of Duplicates**

- Duplicate records can disturb analysis results so are removed

```python
df.duplicated().sum()
```

- The dataset was examined for duplicate records using the duplicated() method.
- No duplicate rows were found in the dataset, indicating that each employee record is unique.

**9. Column name standardisation**

- Column names are standardised for better readability and consistency

```python
df.columns=(df.columns.str.strip()
                      .str.lower()
                      .str.replace(" ","_"))
```

- All column names were standardized by converting them to lowercase, removing extra spaces and replacing spaces with underscores.
- This ensures consistency and prevents case sensitive errors during analysis.

**10. Category Quality Check**

```python
df["deptid"].value_counts()
```

```python
df["positionid"].nunique()
```

- Frequency and uniqueness checks are performed on department and position identifier columns to understand the distribution of employees across the organization.

**11. Data type correction**

```python
categorical_id_columns=["genderid","deptid","marriedid","empsatisfaction","perfscoreid","positionid"]
df[categorical_id_columns]=df[categorical_id_columns].astype("category")
```

- Several columns in the dataset represent categorical identifiers in numerical way.
- So these are converted to category data type for easier analysis and visualisation.
-  Important identifiers like employee id and manager id are retained as numerical value.

**12. Additional features**

**12.1 Employment status**

```python
df["employment_status"]=np.where(df["termd"]==1,"Terminated","Active")
```

- A new feature representing employee employment status is created using the termination indicator.
- This transformation improves interpretability by replacing encoded values with meaningful labels.

**12.2 Annual Salary**

```python
df["annual_salary"]=df["salary"]*12
```

- A new column is created for the annual salary for easier understanding of salaries from different domain.

**12.3 Absence category**

```python
df["absence_level"]=pd.cut(df["absences"],bins=[-1,5,15,df["absences"].max()],labels=["low","medium","high"])
```

- Employee absences are grouped into meaningful categories.
- This step simplifies attendence analysis and enables clearer comparisons across employee groups.

**12.4 Salary Band**

```python
df["salary_band"]=pd.cut(df["salary"],
                         bins=[0,40000,60000,80000,df["salary"].max()],
                         labels=["low","medium","high","very high"])
```

- Salart values are grouped into bands to support categorical salary analysis and improve interpretability.

**12.5 Tenure Calculation**

```python
df["dateofhire"]=pd.to_datetime(df["dateofhire"])
```

'dateofhire' column is converted to datetime format to support time based analysis

```python
df["tenure_years"]=(
    (pd.Timestamp.today() - df["dateofhire"]).dt.days/365).round(1)
```

- Employee tenure is calculated in years to enable analysis of experience related performance and attrition trends

**12.6 Tenure Validation**

```python
df[df["tenure_years"]<0]
```

- Logical validation checks are performed to ensure tenure values are valid and non negative.

**12. Final Dataset Validation**

```python
df.info()
```

```python
df.isnull().sum()
```

- The fully preprocessed dataset was validated using structure, null-value checks, and sample inspection to ensure readiness 
for exploratory data analysis.
- Logical missing values in the termination date column were retained.
- The final cleaned dataset can be exported for exploratory data analysis.

```python
df.to_csv("main_project_preprocessing.csv", index=False)
```

- This step saves the fully cleaned and preprocessed dataset into a csv file for further analysis.

**TASK 2 OUTCOME**
- Effective data cleaning and preprocessing improved data quality, interpretability and analytical readiness ensuring reliable insights in subsequent analysis.

**1.1- Employee Distribution by Employment Status**

**2- BIVARIATE ANALYSIS**
- To examine relationships between two variables and identify influencing factors.

**2.1- Attrition by Department(Groupby Analysis)**

```python
df.groupby("department")["employmentstatus"].value_counts()
```

- The analysis shows that Production has the highest number of both active and terminated employees.
- Voluntary terminations prevail than terminated for cause in most departments, indicating terminations are based on personal or external factors rather than disciplinary issues.

**2.2- Performance Score vs Employee Satisfaction**

- Employee satisfaction was scored as a categorical variable to preserve its original meaning.
- For statistical aggregation, the categorical values were temporarily converted to numeric form to find the mean.

```python
df["empsatisfaction_num"]=df["empsatisfaction"].astype(int)
df.groupby("perfscoreid",observed=True)["empsatisfaction_num"].mean()
```

- Higher performance scores are associated with higher average employee satisfaction, indicating a strong positive relationship between performance and engagement.

**2.3- Average Absences by Employment Status**

```python
df.groupby("employmentstatus")["absences"].mean()
```

- Terminated employees show slightly higher average absences, indicating attendance may influence attrition.

**3- MULTIVARIATE ANALYSIS**
- To analyze relationships involving multiple variables simultaneously.

**3.1- Pivot Table: Department and Employment Status**

```python
pd.pivot_table(df,values="absences",index="department",
               columns="employmentstatus",aggfunc="mean")
```

- The analysis suggests that higher absenteeism is associated with employee termination, highlighting the importance of attendance.
- Absence pattern vary significantly across departments.

**3.2- Salary vs Performance vs Employment Status**

- Since the "perfscoreid" was taken as a categorical attribute, just to avoid future issues the "observed=True" parameter is used in the pivot table.

```python
pd.pivot_table(df,values="salary",index="perfscoreid",
               columns="employmentstatus",aggfunc="mean",observed=True)
```

- Employees with higher performance scores tend to earn higher salaries and are more likely to remain active.

**4- CORRELATION ANALYSIS**
- To measure relationships among numerical variables

```python
df[["salary","absences"]].corr()
```

- A very weak positive correlation indicates that salary has minimal influence on employee absences.

**KEY INSIGHTS FROM EDA**
- Attrition varies by department and performance score.
- Higher satisfaction and performance are linked to employee retention.
- Attendance patterns show a relatioship with employment status.
- Salary aligns closely with performance ratings

## **VISUALIZATIONS**

**OBJECTIVE**
- To visually analyze employee demographics, performance, salary distribution, absenteeism and attrition patterns using appropriate charts for better insights and interpretation.

In this step we import necessary libraries to do visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**1. UNIVARIATE ANALYSIS**

**1.1- Salary Distribution (Histogram)**

```python
plt.figure(figsize=(8,5))
sns.histplot(df["salary"],bins=20,kde=True)
plt.title("Distribution of Employee Salaries")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

- KDE= Kernel Density Estimate

- The histogram shows the distribution of employee salaries. Most employees fall within the lower to mid salary range, indicating a right skewed distribution with fewer high salary employees.

**1.2- Absence Distribution (Histogram)**

```python
plt.figure(figsize=(8,5))
sns.histplot(df["absences"],bins=20,kde=True)
plt.title("Distribution of Employee Absences")
plt.xlabel("Number of Absences")
plt.ylabel("Employee Count")
plt.show()
```

- Majority of the employees have low absence counts, suggesting overall good attendance behavior.

**2. BIVARIATE ANALYSIS**

**2.1- Salary vs Absences (Scatter Plot)**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="salary",y="absences",data=df)
plt.title("Salary vs Absences")
plt.xlabel("Salary")
plt.ylabel("Absences")
plt.show()
```

- The scatter plot shows no clear trend between salary and absences, supporting the earlier correlation result that salary has minimal influence on employee absenteeism.

**2.2- Average Salary by Department (Bar Plot)**

```python
plt.figure(figsize=(10,5))
sns.barplot(x="department",y="salary",data=df,estimator="mean")
plt.xticks(rotation=45)
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

- Salary levels vary across departments with executive and technical roles having higher average compensation. 

**2.3- Absence vs Performance Score (Box Plot)**

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="perfscoreid",y="absences",data=df)
plt.title("Absences Across Performance Scores")
plt.xlabel("Performance Score")
plt.ylabel("Number of Absences")
plt.show()
```

- The box plot indicates that absences vary across performance scores, but no strong or consistent relationship is observed between attendance and performance level.

**3. CATEGORICAL ANALYSIS**

**3.1 Employment Status Distribution (Pie Chart)**

```python
df["employmentstatus"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6),
    title="Employment Status Distribution")
plt.ylabel("")
plt.show()
```

**3.3- Recruitment Source Contribution (Pie Chart)**

**4. MULTIVARIATE ANALYSIS**

**4.1- Salary vs Performance Score (Box Plot)**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="perfscoreid",y="salary",data=df)
plt.title("Salary Distribution by Performance Score")
plt.xlabel("Performance Score")
plt.ylabel("Salary")
plt.show()
```
